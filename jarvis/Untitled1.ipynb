```python
import JarvisAI
import os
import re
import pprint
import random
import warnings
import pickle
import pandas as pd
warnings.filterwarnings("ignore")
warnings.warn("second example of warning!")

obj = JarvisAI.JarvisAssistant()

filename = 'new_model.pkl'
clf = pickle.load(open(filename, 'rb'))
cv=pickle.load(open('new_transform.pkl','rb'))
dataset=pd.read_excel('Disease.xlsx')
filename = 'new_model.pkl'
clf = pickle.load(open(filename, 'rb'))
cv=pickle.load(open('new_transform.pkl','rb'))
disease=""
index=-1
def t2s(text):
    obj.text2speech(text)


def start():
    while True:
        print("Say your AI name to activate")
        status, command = obj.hot_word_detect()
        if status:
            while True:
                # use any one of them
                print("Continue listening, say- 'stop listening to stop continue listening'")
                res = obj.mic_input()
                # res = obj.mic_input_ai(debug=True)
                print(res)

                if re.search("jokes|joke|Jokes|Joke", res):
                    joke_ = obj.tell_me_joke('en', 'neutral')
                    print(joke_)
                    t2s(joke_)
                if re.search('symptoms', res):
                    symptoms=str(res)
                    print(symptoms)
                    print(type(symptoms))
                    data = [symptoms]
                    vect = cv.transform(data).toarray()
                    pred=clf.predict(vect)
                    disease=pred[0]
                    for i in range(0,len(dataset),1):
                        if dataset['Name'][i]==pred[0]:
                            index=i                
                    print('Thank you for your response. You are probably suffering from {}{}. In order to confirm you can know about the causes of the disease just by saying "causes" else you can discover about its diagnosis and treatment by asking'.format(pred[0],dataset['Overview'][index]))
                    t2s('Thank you for your response. You are probably suffering from {}{}. In order to confirm you can know about the causes of the disease just by saying "causes" else you can discover about its diagnosis and treatment by asking'.format(pred[0],dataset['Overview'][index]))

                if re.search('setup|set up', res):
                    setup = obj.setup()
                    print(setup)


                if re.search('weather|temperature', res):
                    city = res.split(' ')[-1]
                    weather_res = obj.weather(city=city)
                    print(weather_res)
                    t2s(weather_res)


                if re.search('tell me about', res):
                    topic = res[14:]
                    wiki_res = obj.tell_me(topic, sentences=1)
                    print(wiki_res)
                    t2s(wiki_res)

                if re.search('date', res):
                    date = obj.tell_me_date()
                    print(date)
                    print(t2s(date))

                if re.search('time', res):
                    time = obj.tell_me_time()
                    print(time)
                    t2s(time)
                
                if re.search("causes|cause", res):
                    if index!=-1:
                        print(dataset['Causes'][index])
                        t2s(dataset['Causes'][index])
                
                                   
                if re.search("diagnosis|detect|identify", res):
                    if (index!=-1):
                        print(dataset['Diagnosis'][index])
                        t2s(dataset['Diagnosis'][index])    
                if re.search("treat|treament|fix", res):
                    if index!=-1:
                        print(dataset['Treatment'][index])
                        t2s(dataset['Treatment'][index])



                if re.search('hello|hi', res):
                    print('Hi')
                    t2s('Hi')

                if re.search('how are you', res):
                    li = ['good', 'fine', 'great']
                    response = random.choice(li)
                    print(f"I am {response}")
                    t2s(f"I am {response}")

                if re.search('your name|who are you', res):
                    print("I am your AI Doctor named JARVIS")
                    t2s("I am your AI Doctor named JARVIS")
                
                if re.search('shit|useless|incompotent|disgusting|garbage|failure', res):
                    
                    print("If you have confusion or the bot is not functioning properly you can visit our site AIDoctor.com")
                    t2s('If you have confusion or the bot is not functioning properly you can visit our site AIDoctor.com')

                if re.search('what can you do', res):
                    
                    ans = """I can help in many ways. I am capable of listening symptoms that you are feeling and predict 
                            the cause of it. Further I can also suggest you how to deal with it.Start by saying jarvis record the symptoms of"""
                    print(ans)
                    t2s(ans)

                if re.search("stop listening|stop", res):
                    break

        else:
            continue


if __name__ == "__main__":
    if not os.path.exists("config/config.ini"):
        res = obj.setup()
        if res:
            print("Settings Saved. Restart your Assistant")
    else:
        start()
```
